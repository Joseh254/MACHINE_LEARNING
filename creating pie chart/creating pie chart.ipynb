```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('C:\\Users\\Thinkpad\\Desktop\\PYTHON\\Machine learning\\music.csv')
data
```

```python
genre_count = data['genre'].value_counts()
# counting types of genres in the dataset
```

```python
# create pie chart and its elements including names of the genretypes and how it will apear using the %1.1ff
plt.pie(genre_count, labels=['dancehall','jazz','classical','dance','acoustic'],autopct='%1.1f%%', startangle=140)
plt.show()
```
